```python
import matplotlib.pyplot as plt
import statsmodels.api as sm
import pandas as pd
import numpy as np
import seaborn as sns
import scipy.stats as stats
import csv
import datetime
```

```python
df = pd.read_csv('dosa.csv')
df.set_index('date')
df.columns
```

```python
df[['quantity','Moving average( 3 days )']].plot(figsize=(20,8))
plt.title('dosa Moving Averages 3 days')
plt.axis('tight')
plt.grid(True)
```

```python
df[['quantity','Moving average( 6 days )']].plot(figsize=(20,8))
plt.title('dosa Moving Average')
plt.axis('tight')
plt.grid(True)
```

```python
df[['quantity','weighted moving average']].plot(figsize=(20,8))
plt.title('dosa weighted Moving Averages 3 days')
plt.axis('tight')
plt.grid(True)
```

```python
df[['quantity', 'exponential smoothing']].plot(figsize=(20,8))
plt.title('dosa exponential smoothing')
plt.axis('tight')
plt.grid(True)
```

```python
df[['quantity',"Holt's model ( Ft )"]].plot(figsize=(20,8))
plt.title('dosa Holts model')
plt.axis('tight')
plt.grid(True)
```

```python
df[['quantity',"Winter's model ( F )"]].plot(figsize=(20,8))
plt.title('dosa Winters model')
plt.axis('tight')
plt.grid(True)
```
