```python
import matplotlib.pyplot as plt
import statsmodels.api as sm
import pandas as pd
import numpy as np
import seaborn as sns
import scipy.stats as stats
import csv
import datetime
```

```python
df = pd.read_csv('vfr.csv')
df.set_index('date')
df.columns
```
